```python
#Didapatkan data okupasi penumpang bus di terminal leuwi panjang, saat ini akan dilakukan analisa trend ramai penumpang
#perilaku penumpang,
# initialization data
import pandas as pd
data_raw = pd.read_csv('DATA/LeuwiPanjang.csv', delimiter = ',', encoding = "ISO-8859-1")
data_raw
```

```python
#PREPROCESSING
data_penumpang=data_raw.iloc[0:34]
data_penumpang_pertahun = data_penumpang.groupby(['TAHUN','BULAN']).mean()
data_penumpang_pertahun.sort_values('NO')
```

```python
#EXTRACTING DATA
#SIKLUS BIS
data_penumpang = data_penumpang_pertahun[['PENUMPANG MASUK','PENUMPANG KELUAR']]
data_penumpang
# data_penumpang_thn2012 = data_penumpang.iloc[0:12]
# data_penumpang_thn2013 = data_penumpang.iloc[12:24]
# data_penumpang_thn2014 = data_penumpang.iloc[24:39] #tidak lengkap hanya hingga oktober
```

```python
#Grafis data penumpang
ax1 = data_penumpang
ax1 = ax1.plot(kind='bar', title ="Data Penumpang Masuk/Keluar Terminal Leuwi Panjang", figsize=(15,3), legend=True, fontsize=12)
ax1.xaxis = data_penumpang['BULAN']
ax1.set_xlabel("BULAN", fontsize=12)
ax1.set_ylabel("Penumpang", fontsize=12)
```

```python
data_penumpang_perbulan = data_penumpang.groupby(['BULAN']).mean()
data_penumpang_perbulan.sort_values('NO')

ax1 = data_penumpang_perbulan
ax1 = ax1.plot(kind='line', title ="Data Penumpang Masuk/Keluar Terminal Leuwi Panjang", figsize=(15,3), legend=True, fontsize=12)
ax1.set_xlabel("BULAN", fontsize=12)
ax1.set_ylabel("BIS MASUK", fontsize=12)
ax1.set_ylabel("BIS KELUAR", fontsize=12)
```
